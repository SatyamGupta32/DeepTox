```python
#MCSHBAP - map, chache, shuffle, batch, prefetch  from_tensor_slices, list_file
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text, y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8)
```

# 2.Creating Sequential Model

```python
# from tensorflow.keras.models import Sequential
# from tensorflow.keras.layers import LSTM, Dropout, Bidirectional,Dense, Embedding

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional, Dense, Embedding
```

```python
y[0]
```

```python
model = Sequential()

# Create the embedding layer 
model.add(Embedding(MAX_FEATURES+1, 32))

# Bidirectional LSTM Layer
model.add(Bidirectional(LSTM(32, activation='tanh')))

# Feature extractor Fully connected layers
model.add(Dense(128, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))

# Build the model by specifying the input shape (batch size is None, sequence length is 100)
model.build(input_shape=(None, 100))

# Final layer 
model.add(Dense(6, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.summary()
```

```python
history = model.fit(train, epochs=2, validation_data=val)
#train the model using the train dataset for one epoch
```

# 4.Evaluate Model

```python

from tensorflow.keras.metrics import Precision, Recall, CategoricalAccuracy
```

```python
pre = Precision()
re = Recall()
acc = CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    # Unpack the batch 
    X_true, y_true = batch
    # Make a prediction 
    yhat = model.predict(X_true)
    
    # Flatten the predictions
    y_true = y_true.flatten()
    yhat = yhat.flatten()
    
    pre.update_state(y_true, yhat)
    re.update_state(y_true, yhat)
    acc.update_state(y_true, yhat)
```

```python
print(f'Precision: {pre.result().numpy()}, Recall:{re.result().numpy()}, Accuracy:{acc.result().numpy()}')
```

```python
import gradio as gr
import speech_recognition as sr

# Function to convert audio to text
def audio_to_text(audio_file):
    recognizer = sr.Recognizer()
    try:
        with sr.AudioFile(audio_file) as source:
            audio = recognizer.record(source)
        text = recognizer.recognize_google(audio)
        return text
    except sr.UnknownValueError:
        return "Audio not understood"
    except sr.RequestError:
        return "Could not request results from Google Speech Recognition service"
    except Exception as e:
        return f"Error processing audio: {str(e)}"

# Function to score comment or audio and return the transcription
def score_comment_or_audio(comment, audio_file):
    # If audio is provided, convert it to text
    transcription = ""
    if audio_file is not None:
        transcription = audio_to_text(audio_file)
        comment = transcription
    
    # Check if we have a valid comment (text or transcribed audio)
    if not comment:
        return "No valid input provided", "No transcription available"
    
    # Use the existing model for toxicity detection
    vectorized_comment = vectorizer([comment])
    results = model.predict(vectorized_comment)
    
    # Format the results
    result_text = ''
    for idx, col in enumerate(df.columns[2:]):
        result_text += '{}: {}\n'.format(col, results[0][idx] > 0.4)
    
    return result_text, transcription

# Create Gradio Interface with both Textbox and Audio inputs
interface = gr.Interface(
    fn=score_comment_or_audio, 
    inputs=[
        gr.Textbox(lines=2, placeholder='Type a comment here'),  # Text input
        gr.Audio(type="filepath")  # Audio input
    ],
    outputs=[
        "text",  # Toxicity score output
        "text"   # Transcription of the audio
    ],
#     live=True  # Enable real-time display
)

interface.launch(share=True)
```
